```python
stats = pd.read_csv(path+'/model_output/traffic/DPG_5v5/statistics.csv').set_index('episode')
stats.head()
```

```python
#collision for agent0
stats.iloc[::20,:10].sum(axis=1).plot(alpha=0.5);
plt.ylabel('collision');
plt.title('Total collision in an episode');
```

```python
#reward for agent0
stats.iloc[::20,10:20].sum(axis=1).plot(alpha=0.5);
plt.ylabel('reward');
plt.title('Total reward in an episode');
```

```python
#loss for agent0
stats.iloc[::20,20:30].mean(axis=1).plot(alpha=0.5);
plt.ylabel('loss');
plt.title('Mean loss (DPG) in an episode');
```
